```python
import numpy as np

def sum_of_integers(list):
  '''Given a list of integers as input,
  calculates the sum of all the elements.'''

  sum_list = list
  print(sum(sum_list))

def min_max(list):
  '''Given list of integers as input
  and finds the maximum and minimum values.'''

  min = list[0]
  max = list[1]

  for i in list:
    if i >= max:
      max = i
    if i <= min:
      min = i

  print ("List minimum is", min)
  print ("List maximum is", max)

def element_sum(list1, list2):
  '''Given two lists of integers,
  calculates element-wise sum.'''

  sums = []
  list1_len = len(list1)
  list2_len = len(list2)

  if list1_len == list2_len:
    print("These lists are equal length")
    sums = [list1[i] + list2[i] for i in range(list1_len)]
    print("Sums are", sums)

def dedupe(list):
  '''Given list of integers, removes all duplicates.'''

  main_list = [list[0]]

  for i in list:
    if i not in main_list:
      main_list.append(i)

  print(main_list)

def ascend_order(list):
  '''Given list of integers, sorts all elements in ascenting order.'''

  ascend = list
  ascend.sort()

  print(ascend)


def non_decreasing(list):
  '''Given a list of integers, checks if the list
  is sorted in non-decreasing order.'''

  check = sorted(list,reverse=True)

  if list == check:
      print("This list is in decreasing order. List order: ",check)
  else:
    print("This list is not in decreasing order")

def list_avg(list):
  '''Given list of integers,
  calculates the average (mean) of all the elements.'''

  total = sum(list)
  count = len(list)
  result = total / count

  print ("The list average is:",result)

def ascend_order_numpy(list):
  '''Given list of integers, sorts all elements in ascenting order.'''

  ascend = list
  array = np.array(ascend.sort())

  print(array)


def non_decreasing_numpy(list):
  '''Given a list of integers, checks if the list
  is sorted in non-decreasing order.'''

  check = np.array(list)
  check_array = sorted(check,reverse=True)

  if list == check_array:
      print("This list is in decreasing order. List order: ",check_array)
  else:
    print("This list is not in decreasing order")

def list_avg_numpy(list):
  '''Given list of integers,
  calculates the average (mean) of all the elements.'''

  array = np.array(list)
  total = sum(array)
  count = len(array)
  result = total / count

  print ("The list average is:",result)


if __name__ == "__main__":
  print("Sum of List Integers:")
  list = [3, 9, 2, 6, 1, 8, 5]
  sum_of_integers(list)

  list = [4, 2, 9, 5, 1, 6, 8]
  sum_of_integers(list)

  list = [1, 2, 3, 4, 5]
  sum_of_integers(list)

  list = [5, 4, 3, 2, 1]
  sum_of_integers(list)

  print ("Minimum number in Lists:")
  list = [3, 9, 2, 6, 1, 8, 5]
  min_max(list)

  list = [4, 2, 9, 5, 1, 6, 8]
  min_max(list)

  list = [1, 2, 3, 4, 5]
  min_max(list)

  list = [5, 4, 3, 2, 1]
  min_max(list)

  print ("The element-wise sums of List1 && List 2 are:")
  List1 = [3, 9, 2, 6, 1, 8, 5]
  List2 = [4, 2, 9, 5, 1, 6, 8]
  element_sum(List1, List2)

  print ("Deduplicated List is:")
  list = [3, 9, 2, 6, 1, 8, 5]
  dedupe(list)

  print ("Ascending ordered list is:")
  list = [4, 2, 9, 5, 1, 6, 8]
  ascend_order(list)

  print ("Descending order check:")
  list = [5, 4, 3, 2, 1]
  non_decreasing(list)

  print ("Average for List:")
  list = [5, 4, 3, 2, 1]
  list_avg(list)

  print ("Ascending ordered array is:")
  list = [4, 2, 9, 5, 1, 6, 8]
  ascend_order_numpy(list)

  print ("Descending order for array check:")
  list = [1, 2, 3, 4, 5]
  non_decreasing_numpy(list)

  print ("Average for Array:")
  list = [5, 4, 3, 2, 1]
  list_avg_numpy(list)
```
